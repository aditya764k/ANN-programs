```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5])
y=np.array([2,4,5,4,5])

m=0.0
b=0.0
learn_rate=0.01
iteration=15


n=len(x)

for i in range(iteration):
    y_predict= m * x + b
    error= y- y_predict
    
    dm = (-2/n) * np.sum(x * error)
    db =(-2/n) * np.sum(error)

    
    m = m - learn_rate*dm
    b = b - learn_rate*db
    
    
    cost = np.mean(error**2)
    print(f"Epoch {i}, Cost: {cost:.4f}, m: {m :.4f}, b: {b :.4f}")
    
    plt.scatter(x,y,color="red",label="actual data")
    plt.plot(x,m*x+b,color="purple",label="predicted")
    plt.title("Linear Regression with Gradient Decent")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.show()

    
    
print(f"Final m: {m: .4f}, Final b: {b:.4f}")

print("final graph")

plt.scatter(x,y,color="red",label="actual data")
plt.plot(x,m*x+b,color="purple",label="predicted")
plt.title("Linear Regression with Gradient Decent")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
